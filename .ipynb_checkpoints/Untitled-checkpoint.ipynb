```python
from pyspark.sql import SparkSession

# Create a Spark session
spark = SparkSession.builder.appName("BDA").getOrCreate()

# Specify the path to your CSV file
csv_file_path = "hdfs://localhost:9820/user/BDAcsv/website_analysis_output_edtech.csv"

# Read the CSV file into a DataFrame
df = spark.read.csv(csv_file_path, header=True, inferSchema=True)

# Convert PySpark DataFrame to Pandas DataFrame
pandas_df = df.toPandas()

# Print the Pandas DataFrame
print("Pandas DataFrame:")
print(pandas_df)

# Describe the Pandas DataFrame
print("\nDescriptive Statistics for Pandas DataFrame:")
print(pandas_df.describe())

# Stop the Spark session
spark.stop()
```
